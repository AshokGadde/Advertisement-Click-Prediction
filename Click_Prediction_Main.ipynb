# Base Models

Test And Train

```python
X = df.drop(labels=['Ad Topic Line','City','Country','Timestamp','Clicked on Ad'], axis=1)
Y = df['Clicked on Ad']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs', max_iter=1000)
model.fit(X_train, Y_train)
```

```python
prediction_Logistic = model.predict(X_test)
prediction_Logistic
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
Y_pred = model.predict(X_test)
log_report = classification_report(Y_test, Y_pred)
print(log_report)
```

```python
cmm= confusion_matrix(Y_test, Y_pred)
sns.heatmap(cmm, annot=True)
```

# Naive Base

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train, Y_train)
prediction_Naive = model.predict(X_test)
prediction_Naive
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
Y_pred = model.predict(X_test)
log_report = classification_report(Y_test, Y_pred)
print(log_report)
```

```python
cmm= confusion_matrix(Y_test, Y_pred)
sns.heatmap(cmm, annot=True)
```

# K Neighbors Classification

```python
from sklearn.neighbors import KNeighborsClassifier
import math
import numpy as np
no_of_neighbours = np.sqrt(len(X_train))
no_of_neighbours = math.floor(no_of_neighbours)
model = KNeighborsClassifier(n_neighbors=no_of_neighbours)
model.fit(X_train, Y_train)
```

```python
prediction_KNN = model.predict(X_test)
prediction_KNN
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
Y_pred = model.predict(X_test)
log_report = classification_report(Y_test, Y_pred)
print(log_report)
```

```python
cmm= confusion_matrix(Y_test, Y_pred)
sns.heatmap(cmm, annot=True)
```
